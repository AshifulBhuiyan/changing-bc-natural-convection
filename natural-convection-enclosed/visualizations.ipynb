```python
import numpy as np
import matplotlib.pyplot as plt
import os

def visualize_csv_field(filepath, field_label="θ", cmap="hot", aspect_ratio=1.0):
    # Load the field
    field = np.loadtxt(filepath, delimiter=",")

    # Infer grid size
    NY, NX = field.shape
    x = np.linspace(0, 1, NX)
    y = np.linspace(0, aspect_ratio, NY)
    X, Y = np.meshgrid(x, y)

    # Plot
    plt.figure(figsize=(8, 6))
    contour = plt.contourf(X, Y, field, levels=20, cmap=cmap)
    plt.colorbar(contour, label=field_label)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"{field_label} field from: {os.path.basename(filepath)}")
    plt.gca().set_aspect('equal')
    plt.tight_layout()
    plt.show()

# --- Example Usage ---
if __name__ == "__main__":
    # Example file path
    filepath = "results/raw_fields/theta_AR1_Ra100_Angle90.csv"
    visualize_csv_field(filepath, field_label="θ", cmap="hot", aspect_ratio=1.0)
```
